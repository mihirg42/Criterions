```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
noise = np.random.normal(0, 0.5, 100)
plt.scatter(range(100), noise)

plt.show()
```

```python
hundred_xs=np.random.uniform(0, 10, 100)

x_y_pairs = []
for x in hundred_xs:
    y  = 22 + 1*(x**(1/2))
    x_y_pairs.append((x,y))
    
xs = []
ys = []
for a,b in x_y_pairs:
    xs.append(a)
    ys.append(b)
    
plt.scatter(xs, ys)
plt.title('Original data')
plt.show()
```

```python
hundred_xs=np.random.uniform(0, 10, 100)

x_y_pairs = []
pos = 0
for x in hundred_xs:
    y  = 22 + 1*(x**(1/2)) + noise[pos]
    x_y_pairs.append((x,y))
    pos+=1
    
xs = []
ys = []
for a,b in x_y_pairs:
    xs.append(a)
    ys.append(b)
    
plt.scatter(xs, ys)
plt.title('Original data')
plt.show()
```
